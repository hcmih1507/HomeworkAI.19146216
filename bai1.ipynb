```python
from os import listdir    # goi he dieu hanh OS 
from numpy import asarray 
from numpy import save 
from keras.utils import load_img, img_to_array 
```

```python
folder = '/content/drive/MyDrive/baocaoai/longbantay/'
photos, labels = list(), list()
for file in listdir(folder):
  output = 0.0
  if file.startswith('van-hinh-dao'):
    output = 1.0
  if file.startswith('van-hinh-chu-thap'):
    output = 2.0
  if file.startswith('van-hinh-tam-giac'):
    output = 3.0
  if file.startswith('van-hinh-vuong'):
    output = 4.0
  if file.startswith('van-hinh-ke-o'):
    output = 5.0
  if file.startswith('van-hinh-ngoi-sao'):
    output = 6.0      
  photo = load_img(folder+file)
  photo = img_to_array(photo)
  photos.append(photo)
  labels.append(output)
photos = asarray(photos)
labels = asarray(labels)
print(photos.shape, labels.shape)
save('/content/drive/MyDrive/baocaoai/handle/minphoto', photos)
save('/content/drive/MyDrive/baocaoai/handle/minlabels', labels)
```
