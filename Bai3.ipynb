```python
from os import listdir    # goi he dieu hanh OS 
from numpy import asarray 
from numpy import save 
from keras.utils import load_img, img_to_array 
```

```python
folder = '/content/drive/MyDrive/baocaoai/thucanVN/'
photos, labels = list(), list()
for file in listdir(folder):
  output = 0.0
  if file.startswith('banhcanh'):
    output = 1.0
  if file.startswith('banhxeo'):
    output = 2.0
  if file.startswith('bunbo'):
    output = 3.0
  if file.startswith('che'):
    output = 4.0
  if file.startswith('comtam'):
    output = 5.0
  if file.startswith('hutieu'):
    output = 6.0    
  if file.startswith('myquang'):
    output = 7.0  
  if file.startswith('pho'):
    output = 8.0  
  if file.startswith('tauhu'):
    output = 9.0  
  if file.startswith('xoi'):
    output = 10.0      
  photo = load_img(folder+file)
  photo = img_to_array(photo)
  photos.append(photo)
  labels.append(output)
photos = asarray(photos)
labels = asarray(labels)
print(photos.shape, labels.shape)
save('/content/drive/MyDrive/baocaoai/handle/min3photo', photos)
save('/content/drive/MyDrive/baocaoai/handle/min3labels', labels)
```
