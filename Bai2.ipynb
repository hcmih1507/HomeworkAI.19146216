```python
from os import listdir    # goi he dieu hanh OS 
from numpy import asarray 
from numpy import save 
from keras.utils import load_img, img_to_array 
```

```python
folder = '/content/drive/MyDrive/baocaoai/tiengiayvnd/'
photos, labels = list(), list()
for file in listdir(folder):
  output = 0.0
  if file.startswith('namk'):
    output = 1.0
  if file.startswith('muoik'):
    output = 2.0
  if file.startswith('haimuoik'):
    output = 3.0
  if file.startswith('nammuoik'):
    output = 4.0
  if file.startswith('mottramk'):
    output = 5.0
  if file.startswith('haitramk'):
    output = 6.0     
  if file.startswith('namtramk'):
    output = 7.0      
  photo = load_img(folder+file)
  photo = img_to_array(photo)
  photos.append(photo)
  labels.append(output)
photos = asarray(photos)
labels = asarray(labels)
print(photos.shape, labels.shape)
save('/content/drive/MyDrive/baocaoai/handle/min2photo', photos)
save('/content/drive/MyDrive/baocaoai/handle/min2labels', labels)
```
