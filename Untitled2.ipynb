```python
import numpy as np 
import skfuzzy as fuzz 
from skfuzzy import control as  ctrl 
 
```

```python
food=ctrl.Antecedent(np.arange(0,11,1),'food')
service=ctrl.Antecedent(np.arange(0,11,1),'service')
tip=ctrl.Consequent(np.arange(10,31,1),'tip')
food['poor']=fuzz.trimf(food.universe,[0,0,5])
food['average']=fuzz.trimf(food.universe,[2,5,7])
food['good']=fuzz.trimf(food.universe,[5,8,10])

service['poor']=fuzz.trimf(service.universe,[0,0,5])
service['average']=fuzz.trimf(service.universe,[2,5,7])
service['good']=fuzz.trimf(service.universe,[5,8,10])

tip['less']=fuzz.trimf(tip.universe,[10,15,20])
tip['normal']=fuzz.trimf(tip.universe,[15,20,25])
tip['much']=fuzz.trimf(tip.universe,[20,25,30])
food.view()
service.view()
tip.view()
food['good'].view()
```

```python
rule1=ctrl.Rule(food['poor'] & service['poor'],tip['less'])
rule2=ctrl.Rule(food['poor'] & service['average'],tip['less'])
rule3=ctrl.Rule(food['poor'] & service['good'],tip['less'])
rule4=ctrl.Rule(food['average'] & service['poor'],tip['less'])
rule5=ctrl.Rule(food['average'] & service['average'],tip['less'])
rule6=ctrl.Rule(food['average'] & service['good'],tip['normal'])
rule7=ctrl.Rule(food['good'] & service['poor'],tip['less'])
rule8=ctrl.Rule(food['good'] & service['average'],tip['normal'])
rule9=ctrl.Rule(food['good'] & service['good'],tip['much'])
tipping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['food']=6.5
tipping.input['service']=9.8
tipping.compute()
print(tipping.output['tip'])
tip.view(sim=tipping)
```

```python
import numpy as np 
import skfuzzy as fuzz 
from skfuzzy import control as  ctrl 
```

```python
time=ctrl.Antecedent(np.arange(0,121,1),'time')
rice=ctrl.Antecedent(np.arange(0,1.5,0.1),'rice')
power=ctrl.Consequent(np.arange(0,101,1),'power')
time['veryslow']=fuzz.trimf(time.universe,[0,0,30])
time['slow']=fuzz.trimf(time.universe,[0,30,60])
time['medium']=fuzz.trimf(time.universe,[30,60,90])
time['fast']=fuzz.trimf(time.universe,[60,90,120])
time['veryfast']=fuzz.trimf(time.universe,[90,120,120])

rice['veryfew']=fuzz.trimf(rice.universe,[0.3,0.3,0.6])
rice['few']=fuzz.trimf(rice.universe,[0.3,0.6,0.9])
rice['average']=fuzz.trimf(rice.universe,[0.6,0.9,1.2])
rice['much']=fuzz.trimf(rice.universe,[0.9,1.2,1.5])
rice['verymuch']=fuzz.trimf(rice.universe,[1.2,1.5,1.5])

power['min']=fuzz.trimf(power.universe,[0,0,25])
power['low']=fuzz.trimf(power.universe,[0,25,50])
power['normal']=fuzz.trimf(power.universe,[25,50,75])
power['high']=fuzz.trimf(power.universe,[50,75,100])
power['max']=fuzz.trimf(power.universe,[75,100,100])
time.view()
rice.view()
power.view()
time['medium'].view()
```

```python

rule1=ctrl.Rule(time['veryslow'] & rice['veryfew'],power['min'])
rule2=ctrl.Rule(time['veryslow'] & rice['few'],power['low'])
rule3=ctrl.Rule(time['veryslow'] & rice['average'],power['low'])
rule4=ctrl.Rule(time['veryslow'] & rice['much'],power['normal'])
rule5=ctrl.Rule(time['veryslow'] & rice['verymuch'],power['normal'])

rule6=ctrl.Rule(time['slow'] & rice['veryfew'],power['low'])
rule7=ctrl.Rule(time['slow'] & rice['few'],power['low'])
rule8=ctrl.Rule(time['slow'] & rice['average'],power['normal'])
rule9=ctrl.Rule(time['slow'] & rice['much'],power['normal'])
rule10=ctrl.Rule(time['slow'] & rice['verymuch'],power['normal'])

rule11=ctrl.Rule(time['medium'] & rice['veryfew'],power['low'])
rule12=ctrl.Rule(time['medium'] & rice['few'],power['normal'])
rule13=ctrl.Rule(time['medium'] & rice['average'],power['normal'])
rule14=ctrl.Rule(time['medium'] & rice['much'],power['high'])
rule15=ctrl.Rule(time['medium'] & rice['verymuch'],power['high'])

rule16=ctrl.Rule(time['fast'] & rice['veryfew'],power['normal'])
rule17=ctrl.Rule(time['fast'] & rice['few'],power['normal'])
rule18=ctrl.Rule(time['fast'] & rice['average'],power['high'])
rule19=ctrl.Rule(time['fast'] & rice['much'],power['high'])
rule20=ctrl.Rule(time['fast'] & rice['verymuch'],power['high'])

rule21=ctrl.Rule(time['veryfast'] & rice['veryfew'],power['normal'])
rule22=ctrl.Rule(time['veryfast'] & rice['few'],power['normal'])
rule23=ctrl.Rule(time['veryfast'] & rice['average'],power['high'])
rule24=ctrl.Rule(time['veryfast'] & rice['much'],power['high'])
rule25=ctrl.Rule(time['veryfast'] & rice['verymuch'],power['max'])


powerping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powerping=ctrl.ControlSystemSimulation(powerping_ctrl)
powerping.input['time']=65
powerping.input['rice']=0.8
powerping.compute()
print(powerping.output['power'])
power.view(sim=powerping)
```
